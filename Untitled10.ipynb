Now, let's load the `anime.csv` file into a pandas DataFrame and display its first 5 rows.

```python
import pandas as pd

anime_df = pd.read_csv('anime_data/anime.csv')
print("Anime DataFrame head:")
display(anime_df.head())
```

Next, let's load the `rating.csv` file into a pandas DataFrame and display its first 5 rows.

```python
rating_df = pd.read_csv('anime_data/rating.csv')
print("Rating DataFrame head:")
display(rating_df.head())
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# Handle potential NaN values by filling them with an empty string
anime_df['genre'] = anime_df['genre'].fillna('')

# Initialize CountVectorizer
count_vectorizer = CountVectorizer(tokenizer=lambda x: x.split(', ')) # Assuming genres are comma-separated

# Fit and transform the genre column
genre_matrix = count_vectorizer.fit_transform(anime_df['genre'])

print("Shape of genre matrix:", genre_matrix.shape)
print("Number of unique genres:", len(count_vectorizer.get_feature_names_out()))
print("First 5 feature names (genres):")
print(count_vectorizer.get_feature_names_out()[:5])
```

Now, let's switch to `scikit-surprise` for more advanced recommendation system functionalities. We'll start by installing the library.

```python
!pip install scikit-surprise
```

After installing, we need to load the `rating_df` into a format that `surprise` can understand. This involves defining a `Reader` object that specifies the scale of the ratings and then loading the data into a `Dataset`.

```python
from surprise import Dataset, Reader

# A Reader is needed to parse the file or DataFrame.
# It tells Surprise how to read the data, e.g., the rating scale.
# We are now assuming ratings are strictly between 1 and 10, after filtering out the 0s.
reader = Reader(rating_scale=(1, 10))

# Filter out ratings that are 0 (which were originally -1) as they represent 'no explicit rating'.
filtered_rating_df = rating_df[rating_df['rating'] != -1]

# Load the data from the filtered DataFrame into a Surprise Dataset object
data = Dataset.load_from_df(filtered_rating_df[['user_id', 'anime_id', 'rating']], reader)

print("Data loaded into Surprise dataset format after filtering 0 ratings.", filtered_rating_df.shape)
```

Now that the data is loaded, let's train a recommendation model using the SVD (Singular Value Decomposition) algorithm. We'll also split the data into training and test sets to evaluate the model's performance.

```python
from surprise import SVD
from surprise.model_selection import train_test_split
from surprise import accuracy

# Split the data into 80% training set and 20% test set
trainset, testset = train_test_split(data, test_size=0.2, random_state=42)

# Use the SVD algorithm
algo = SVD(n_factors=100, random_state=42)

# Train the algorithm on the trainset
algo.fit(trainset)

# Predict ratings for the testset
predictions = algo.test(testset)

# Then compute RMSE
rmse = accuracy.rmse(predictions, verbose=False)
print(f"RMSE on test set: {rmse:.4f}")

# Optionally, display some example predictions
print("\nExample Predictions:")
for uid, iid, true_r, est, _ in predictions[:5]:
    print(f"User {uid} rated item {iid} with true rating {true_r:.1f}, predicted rating {est:.1f}")
```

### User-Mean Normalization and SVD

This section implements user-mean normalization on the ratings. This technique adjusts each user's ratings by subtracting their average rating, allowing the SVD model to focus on relative preferences. After the model makes predictions on these normalized values, we will denormalize them to get predictions back in the original rating scale (1-10).

```python
import pandas as pd
import numpy as np
from surprise import Dataset, Reader

# Use filtered_rating_df as the base, which already excludes -1 ratings
# Make a copy to avoid modifying the original DataFrame directly for this experiment
processed_rating_df_normalized = filtered_rating_df.copy()

# Calculate the mean rating for each user, based on their explicit ratings (1-10)
# No need to replace -1 with NaN as filtered_rating_df already excludes them.
user_means_normalized = processed_rating_df_normalized.groupby('user_id')['rating'].mean()

# Store user means in a dictionary for easy lookup.
# Users with no explicit ratings in filtered_rating_df might not exist in user_means_normalized, so .get() with default 0 is safe.
user_means_dict_normalized = user_means_normalized.fillna(0).to_dict()

# Normalize ratings: subtract user mean from explicit ratings
# Since -1 ratings are already filtered out, we only process explicit ratings.
def normalize_rating(row):
    user_id = row['user_id']
    original_rating = row['rating']
    user_mean = user_means_dict_normalized.get(user_id, 0) # Get user's mean (0 if user has no explicit ratings in filtered_df)
    return original_rating - user_mean

processed_rating_df_normalized['normalized_rating'] = processed_rating_df_normalized.apply(normalize_rating, axis=1)

# Determine the rating scale for the Reader based on these normalized ratings.
min_normalized_rating = processed_rating_df_normalized['normalized_rating'].min()
max_normalized_rating = processed_rating_df_normalized['normalized_rating'].max()

# A Reader is needed to parse the DataFrame and specify the scale of the normalized ratings.
reader_normalized = Reader(rating_scale=(min_normalized_rating, max_normalized_rating))

# Load the processed DataFrame into a Surprise Dataset object, using the normalized ratings
data_normalized = Dataset.load_from_df(processed_rating_df_normalized[['user_id', 'anime_id', 'normalized_rating']], reader_normalized)

print(f"Data loaded into Surprise dataset format after user-mean normalization using filtered_rating_df. Shape: {processed_rating_df_normalized.shape}")
print(f"Normalized rating scale for Reader: ({min_normalized_rating:.2f}, {max_normalized_rating:.2f})")
```

```python
from surprise import SVD
from surprise.model_selection import train_test_split
from surprise import accuracy
from surprise import Prediction # Import Prediction object for denormalization

# Split the normalized data into 80% training set and 20% test set
# trainset and testset will contain normalized ratings
trainset_normalized, testset_normalized = train_test_split(data_normalized, test_size=0.2, random_state=42)

# Use the SVD algorithm
algo_normalized = SVD(n_factors=100, random_state=42)

# Train the algorithm on the trainset (which contains normalized ratings)
algo_normalized.fit(trainset_normalized)

# Predict normalized ratings for the testset
predictions_normalized = algo_normalized.test(testset_normalized)

# Function to convert normalized prediction back to original scale (1-10)
def denormalize_prediction(uid, normalized_value):
    # Get user's original mean (0 if no explicit ratings, from user_means_dict_normalized)
    user_mean = user_means_dict_normalized.get(uid, 0)
    # Add the user's mean back to get the value in the original scale
    return normalized_value + user_mean

# Calculate RMSE on the normalized predictions (what SVD was trained to predict)
rmse_on_normalized_data = accuracy.rmse(predictions_normalized, verbose=False)
print(f"RMSE on normalized test set: {rmse_on_normalized_data:.4f}")

# To calculate RMSE on the original rating scale, we need to denormalize both true_r and est.
original_scale_predictions_for_rmse = []
for p in predictions_normalized:
    # The true rating (r_ui) here is still the normalized rating. We need to denormalize it.
    # The estimated rating (est) is also normalized, so we denormalize it.
    original_true_r = denormalize_prediction(p.uid, p.r_ui)
    original_est = denormalize_prediction(p.uid, p.est)

    # Clip denormalized predictions to the original expected range (1 to 10)
    original_est = max(1, min(10, original_est))

    # Create a new Prediction object with denormalized true and estimated ratings
    original_scale_predictions_for_rmse.append(Prediction(uid=p.uid, iid=p.iid, r_ui=original_true_r, est=original_est, details=p.details))

rmse_on_original_scale_after_denorm = accuracy.rmse(original_scale_predictions_for_rmse, verbose=False)
print(f"RMSE on original scale test set (after denormalization and clipping): {rmse_on_original_scale_after_denorm:.4f}")


# Display some example predictions, denormalized to the original scale for interpretability
print("\nExample Predictions (denormalized to original scale):")
for p in predictions_normalized[:5]:
    original_true_r = denormalize_prediction(p.uid, p.r_ui)
    original_est = denormalize_prediction(p.uid, p.est)
    # Clip the displayed estimated rating for better interpretability
    displayed_est = max(1, min(10, original_est))
    print(f"User {p.uid} rated item {p.iid} with true rating {original_true_r:.1f}, predicted rating {displayed_est:.1f}")
```
